```python
import codecs
from bs4 import BeautifulSoup
import time
```

```python
page = open('./html_convert/1.html', 'rt', encoding='utf-8').read() # HTML 파일 읽고 문자열 리턴 
soup = BeautifulSoup(page, 'html.parser') # soup 객체 생성 

# Soup 객체를 이용해서 문서를 출력 
# print(soup.prettify())
```

```python
span_hrt_cs_11 = soup.find_all('span', {'class' : 'hrt cs11'})
length = len(soup.find_all('span', {'class' : 'hrt cs11'}))

text_array = []

for i in range(length) : 
    text = span_hrt_cs_11[i].text
#     print(text)
    
    if text[-1] == '?' : 
        text_array.append(text)
        print(text)
        
    
```

```python
text_array = []

for n in range(1, 27) :
    page = open(f'./html_convert/{n}.html', 'rt', encoding='utf-8').read() # HTML 파일 읽고 문자열 리턴 
    print(f'./html_convert/{n}.html')
    soup = BeautifulSoup(page, 'html.parser') # soup 객체 생성 

    span_hrt_cs_11 = soup.find_all('span', {'class' : 'hrt cs11'})
    length = len(soup.find_all('span', {'class' : 'hrt cs11'}))


    for i in range(length) : 
        text = span_hrt_cs_11[i].text
#         print(text[0])

        try : 
#         text[-1] == '?' and 
            if int(text[0]) in range(1, 10) : 
                text_array.append(text)
                print(text)
        except : 
            pass
    
    time.sleep(3)
```

```python
# 해당 텍스트 전처리해야겠네
preprocessing_text_arr = []

for text in text_array : 
    

    if text[-1] != '?' : 
        if text == '47. 전기 화재시 가장 좋은? 소화기는' : 
            preprocessing_text_arr.append(text.split('. ')[1])
            continue
        
        temp_idx = text.index('?')
#         print(text[temp_idx + 2])
        pre_text = text.split(text[temp_idx + 2])[0][:-1]
        preprocessing_text_arr.append(pre_text.split('. ')[1])
#         print(text)
        
    else : 
        preprocessing_text_arr.append(text.split('. ')[1])
        
print(len(preprocessing_text_arr))
```
