```python
# 스크롤 끝까지 밑으로 내리기 함수 구현

def scroll_ending(browser) : 
    prev_height = browser.excute_script("return document. body.scrollHeight")
    
    while True:
        #첫번째로 스크롤 내리기
        browser.excute_script("window.scrollTo(0,document.body.scrollHeight)")

        #시간대기
        time.sleep(1)

        #현재높이 저장
        current_height = browser.excute_script("return document. body.scrollHeight")

        #현재높이와 끝의 높이가 끝이면 탈출
        if current_height == prev_height:
            break
        #업데이트해줘서 끝낼 수 있도록
        prev_height == current_height
```

```python
import pandas as pd 
import warnings 
from selenium import webdriver
from selenium.webdriver import ActionChains 
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from bs4 import BeautifulSoup
from tqdm import tqdm_notebook
import time
warnings.simplefilter(action = 'ignore')
```

```python
# 1. cbt 사이트 접근 (자격증 기출문제 전자 문제 html)
url = 'https://www.comcbt.com/'
driver = webdriver.Chrome('C:/Users/ck12q/OneDrive/Documents/ds_study/driver/chromedriver.exe')
driver.get(url)
```

```python
# row 115
# col 3
# if row == 6 and col == 111 : break
 
col = 1

for _ in tqdm_notebook(range(3)) :
    
    row = 1
    
    while row <= 115 : 
        
#         print(f'body > center:nth-child(5) > center > table > tbody > tr:nth-child(2) > td:nth-child({col}) > li:nth-child({row}) > a')
    
        driver.find_element(By.CSS_SELECTOR,
                            f'body > center:nth-child(5) > center > table > tbody > tr:nth-child(2) > td:nth-child({col}) > li:nth-child({row}) > a').click()
        
        time.sleep(1)
        
        # 탭 전환
        last_tab = driver.window_handles[-1]
        driver.switch_to.window(window_name=last_tab)
        time.sleep(1)
        
        # 시험지 관련 글에 대한 셀레니움 개체 모임
        writing_category = driver.find_elements(By.CSS_SELECTOR, 'div.bd_lst_wrp > table > tbody > tr')
        
        for i in range(2, len(writing_category) + 1) : 
            
            # 글 위에서부터 하나씩 클릭
            driver.find_element(By.CSS_SELECTOR, f'div.bd_lst_wrp > table > tbody > tr:nth-child({i}) > td.title > a').click()
            time.sleep(1.5)
            print(driver.find_element(By.CSS_SELECTOR, 'div.rd.rd_nav_style2.clear > div.rd_hd.clear > div > div.top_area.ngeb > h1 > a').text)
            try :
                driver.find_element_by_css_selector('div.rd.rd_nav_style2.clear > div.rd_body.clear > article > div > p:nth-child(20) > a').click()
            except NoSuchElementException : 
                pass
            time.sleep(1)
            driver.find_element(By.CSS_SELECTOR, '#column > div > nav > ul > li.li2.on.on2 > ul > li.li3.on.on3').click()
            time.sleep(1)
            
        
        
        
        # 탭 close
        driver.close()
        first_tab = driver.window_handles[0]
        driver.switch_to.window(window_name=first_tab )
        time.sleep(0.5)
        
        
        row += 1 
        
    col += 1

    
```
